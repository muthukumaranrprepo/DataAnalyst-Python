```python
import pandas as pd
datadf = pd.read_csv(r'C:\Users\muthu\OneDrive\Desktop\TCR_Date Total.csv')
datadf.head()
```

```python
def returnsum(data) :
    data1 = data.split('-')
    if data1[1] == '01' :
       return 'Jan' 
    elif data1[1] == '02' :
       return 'Feb'
    if data1[1] == '03' :
       return 'Mar' 
    elif data1[1] == '04' :
       return 'Apr'
    elif data1[1] == '05' :
       return 'May'
    elif data1[1] == '06' :
       return 'Jun'
    elif data1[1] == '07' :
       return 'Jul'
    elif data1[1] == '08' :
       return 'Aug'
    return None

datadf['month'] = datadf['Date'].apply(returnsum)
datadf
```

```python
resultdata = datadf.pivot_table(index='month',values='sum(TotalINvested)',aggfunc=sum)
resultdata
```
