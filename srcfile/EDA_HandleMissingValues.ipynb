```python
# EDA :: Learn how to handle the missing values. Use DS - Churn_Modelling.csv file.

import pandas as pd

churnmodeldf = pd.read_csv('https://github.com/muthukumaranrprepo/DataAnalyst-Python/blob/main/DataFiles/Churn_Modelling.csv?raw=True')
churnmodeldf.info()
```

```python
#check for null values 
churnmodeldf.isnull().sum()
```

```python
#use dropna function to drop the null values
churnmodeldf_up = churnmodeldf.dropna(axis=1)
churnmodeldf_up.info() 
```

```python
# removing the records that has null values
churnmodeldf_up = churnmodeldf.dropna(axis=0)
churnmodeldf_up.info() 
```
