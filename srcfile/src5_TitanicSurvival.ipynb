```python
#Titanic Survival Prediction Using Machine Learning

import pandas as pd 
import matplotlib.pyplot as plt

test_df = pd.read_csv('https://github.com/muthukumaranrprepo/DataAnalyst-Python/blob/main/DataFiles/test.csv?raw=True')
train_df = pd.read_csv('https://github.com/muthukumaranrprepo/DataAnalyst-Python/blob/main/DataFiles/train.csv?raw=True')



```

```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.pointplot(train_df)
```

```python
plt.style.use('fivethirtyeight')
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(10,6))
train_df['Survived'].value_counts().plot.pie(ax=ax1,explode=[0, 0.1], autopct='%1.1f%%', shadow=True)
ax1.set_title('Survivors (1) and the dead (0)') 
ax1.set_ylabel('') 
sns.countplot(x='Survived', data=train_df, ax=ax2)
ax2.set_title('Survivors (1) and the dead (0)') 
ax2.set_ylabel('') 
```

```python
train_df.head()
```

```python
for i, predictor in enumerate(test_df[['Sex','Embarked']]) :
    plt.figure(i)
    sns.countplot(y=predictor, data=train_df,hue='Survived')
    plt.title(f'Survival prediction based on {predictor}')
    plt.show()
```
