```python
#Project :: Microsoft Analysis

import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

salesdf = pd.read_csv('https://github.com/muthukumaranrprepo/DataAnalyst-Python/blob/main/DataFiles/FinanacialSalesSample_MS.csv?raw=True')

salesdf.head()
```

```python
#check any null values
salesdf.isnull().sum()
```

```python
missing = pd.DataFrame(salesdf.isnull().sum() * 100 / len(salesdf)).reset_index()
plt.figure(figsize=(20,4))
sns.pointplot(data=missing, x='index', y=0)
```
