```python
#EDA :: Univarite Analysis for Categorical Columns
for i,predictor in enumerate(cat_col) :
    plt.figure(i)
    sns.countplot(x=fin_df[predictor],data=fin_df,hue='gender')
    plt.title(f'Countplot for {predictor}')
    plt.show()
```

```python
#correlation matrix for numerical columns : 
sns.heatmap(num_type.corr(),annot=True,cmap='coolwarm')
```

```python
fin_df[num_col].hist(bins=15, figsize=(15, 6), layout=(2, 4))
plt.suptitle("Distribution of Investment Preferences")
plt.show()
```
