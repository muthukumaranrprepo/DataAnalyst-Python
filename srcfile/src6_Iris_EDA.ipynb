```python
#project  :: 6 :: Iris Dataset :: EDA

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
data_df = pd.read_csv('https://github.com/muthukumaranrprepo/DataAnalyst-Python/blob/main/DataFiles/Iris.csv?raw=True')
data_df.info()
```

```python
#check for null values
data_df.isnull().sum()
```

```python
# Univarite Analysis
# #having a look at the type of data all are numerical data , except species which is categorical. 
#try ti get unique values in species column and use categorical plot 

data_df['Species'].unique()
```

```python
sns.countplot(x='Species', data=data_df, palette='Set2')
```

```python
#explore bivarite analysis using scatter plot 

plt.Figure(figsize=(10, 6))
fig,(ax1,ax2) = plt.subplots(1, 2, figsize=(12, 6))

sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=data_df, ax=ax1)
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=data_df, ax=ax2)

plt.show()
```

```python
#pairplot - can be used for bivarite analysis
# sns.pairplot (from Seaborn) is used when you want a quick, exploratory overview of relationships among multiple numeric variables in a dataset.
# use hue for categorical variable to color the data points   

sns.pairplot(data_df.drop(['Id'], axis=1), hue='Species', palette='Set2')
```

```python
# another Bi varite Analysis using - Histplot 


fig, axes = plt.subplots(2, 2, figsize=(10,10))

axes[0,0].set_title("Sepal Length")
axes[0,0].hist(data_df['SepalLengthCm'], bins=7)

axes[0,1].set_title("Sepal Width")
axes[0,1].hist(data_df['SepalWidthCm'], bins=5);

axes[1,0].set_title("Petal Length")
axes[1,0].hist(data_df['PetalLengthCm'], bins=6);

axes[1,1].set_title("Petal Width")
axes[1,1].hist(data_df['PetalWidthCm'], bins=6);
```

```python
#Another Bivarite analysis using distplot 

plot = sns.FacetGrid(data_df, hue="Species")
plot.map(sns.kdeplot, "SepalLengthCm").add_legend()

plot = sns.FacetGrid(data_df, hue="Species")
plot.map(sns.kdeplot, "SepalWidthCm").add_legend()
```
