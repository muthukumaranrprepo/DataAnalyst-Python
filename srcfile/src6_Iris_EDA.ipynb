```python
#project  :: 6 :: Iris Dataset :: EDA

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
data_df = pd.read_csv('https://github.com/muthukumaranrprepo/DataAnalyst-Python/blob/main/DataFiles/Iris.csv?raw=True')
data_df.info()
```

```python
#check for null values
data_df.isnull().sum()
```

```python
# Univarite Analysis
# #having a look at the type of data all are numerical data , except species which is categorical. 
#try ti get unique values in species column and use categorical plot 

data_df['Species'].unique()
```

```python
sns.countplot(x='Species', data=data_df, palette='Set2')
```

```python
#explore bivarite analysis using scatter plot 

plt.Figure(figsize=(10, 6))
fig,(ax1,ax2) = plt.subplots(1, 2, figsize=(12, 6))

sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=data_df, ax=ax1)
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=data_df, ax=ax2)

plt.show()
```
