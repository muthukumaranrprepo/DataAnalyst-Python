```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn import preprocessing
datadf = pd.read_csv('https://github.com/muthukumaranrprepo/DataAnalyst-Python/blob/main/DataFiles/Churn_Modelling.csv?raw=True')
datadf.info()
```

```python
datadf.Gender.value_counts()
```

```python
#handle the missing values. Find the most used value with mode()
datadf.Gender.mode()
```

```python
#update the missing values using fillna with value returned by mode()
datadf['Gender'] = datadf.Gender.fillna('Male')
datadf.Gender.value_counts()
```

```python
#Label Encoding Technique
import statistics
from sklearn.preprocessing import LabelEncoder #scikit-learn library ML & statistics model 

le = LabelEncoder()
datadf['Gender_label'] = le.fit_transform(datadf['Gender'].values)
datadf['Gender_label'].value_counts()
```

```python
# newly created column available at end of df 0 - for female and 1 - male 
datadf.head()
```

```python
#One Hot Encoding Technique - sklearn.OneHotEncoder or pd.get_dummies()

datadf_en = pd.get_dummies(datadf['Geography'])
datadf_en.head()
```

```python
#drop surnames and all onehot encoding columns available at end of DF
datadf_new = pd.get_dummies(datadf)
datadf_new.head()
```
